```python
import pandas as pd

swiggy_df = pd.read_csv('C:/Users/Tanavi/OneDrive/Desktop/New folder/PROJECT/swiggy.csv')
```

```python
# Swiggy Dataset
print("Swiggy Data Info:")
print(swiggy_df.info())

print("Swiggy Data Summary:")
print(swiggy_df.describe())
```

```python
# Handle missing values
swiggy_df.dropna(inplace=True)
```

```python
# Convert data types if necessary
swiggy_df['Price'] = swiggy_df['Price'].astype(float)
swiggy_df['Avg ratings'] = swiggy_df['Avg ratings'].astype(float)
```

```python
# Import necessary libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with a specified size for better readability
plt.figure(figsize=(8, 6))

# Plot a boxplot to visualize the distribution of 'Price' with outliers for Swiggy orders
sns.boxplot(x=swiggy_df['Price'])

# Set the title of the boxplot
plt.title("Price Distribution of Swiggy Orders with Outliers", fontsize=14)

# Set the x-axis label for the price of Swiggy orders
plt.xlabel("Price of Swiggy Orders (in INR)", fontsize=12)

# Set the y-axis label for frequency, though boxplots typically don't use frequency on the y-axis
plt.ylabel("Frequency", fontsize=12)

# Display the plot
plt.show()
```

```python
# Average Delivery Time
print("Average Delivery Time (Swiggy):", swiggy_df['Delivery time'].mean())
print("Median Delivery Time (Swiggy):", swiggy_df['Delivery time'].median())
```

```python
# Create a figure with a specified size for better readability
plt.figure(figsize=(8, 6))

# Plot a histogram with Kernel Density Estimation (KDE) to visualize the distribution of 'Avg ratings'
# The 'Avg ratings' column is used, with 10 bins and a blue color
sns.histplot(swiggy_df['Avg ratings'], bins=10, kde=True, color="blue")

# Set the title of the plot
plt.title("Swiggy - Ratings Distribution")

# Set the x-axis label for the average ratings
plt.xlabel("Average Rating")

# Set the y-axis label for frequency of ratings
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
# Plot the frequency of the top 10 most common food types from the 'Food type' column
# The 'Food type' values are counted and plotted as a bar chart with green color
swiggy_df['Food type'].value_counts().head(10).plot(kind='bar', color='green')

# Set the title of the bar plot
plt.title("Most Common Food Types on Swiggy")

# Set the x-axis label for food types
plt.xlabel("Food Type")

# Set the y-axis label for the frequency of orders for each food type
plt.ylabel("Order Frequency")

# Display the plot
plt.show()
```

```python
# Create a figure with a specified size for better readability
plt.figure(figsize=(8, 6))

# Plot a histogram to visualize the distribution of 'Price' with 20 bins and purple color
sns.histplot(swiggy_df['Price'], bins=20, color='purple')

# Set the title of the plot
plt.title("Swiggy - Price Range Distribution")

# Set the x-axis label for price values
plt.xlabel("Price")

# Set the y-axis label for frequency of price values
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
# Import necessary libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for seaborn plots
sns.set(style="whitegrid")

# Load Swiggy dataset from a CSV file into a DataFrame
swiggy_df = pd.read_csv('C:/Users/Tanavi/OneDrive/Desktop/New folder/PROJECT/swiggy.csv')

### Data Cleaning for Swiggy ###

# Handle missing values by dropping rows with any missing data
swiggy_df.dropna(inplace=True)

# Convert 'Price' and 'Avg ratings' columns to float data type for consistency
swiggy_df['Price'] = swiggy_df['Price'].astype(float)
swiggy_df['Avg ratings'] = swiggy_df['Avg ratings'].astype(float)

### Swiggy EDA ###

# Print the average and median delivery times to get insights into delivery speed
print("Average Delivery Time (Swiggy):", swiggy_df['Delivery time'].mean())
print("Median Delivery Time (Swiggy):", swiggy_df['Delivery time'].median())

# Plot the distribution of average ratings using a histogram with KDE (Kernel Density Estimate)
plt.figure(figsize=(8, 6))
sns.histplot(swiggy_df['Avg ratings'], bins=10, kde=True, color="blue")
plt.title("Swiggy - Ratings Distribution")  # Set plot title
plt.xlabel("Average Rating")  # Set x-axis label
plt.ylabel("Frequency")  # Set y-axis label
plt.show()  # Display the plot

# Plot the frequency of the top 10 most common food types in the dataset as a bar chart
swiggy_df['Food type'].value_counts().head(10).plot(kind='bar', color='green')
plt.title("Most Common Food Types on Swiggy")  # Set plot title
plt.xlabel("Food Type")  # Set x-axis label
plt.ylabel("Order Frequency")  # Set y-axis label
plt.show()  # Display the plot

# Plot the distribution of prices using a histogram with 20 bins and purple color
plt.figure(figsize=(8, 6))
sns.histplot(swiggy_df['Price'], bins=20, color='purple')
plt.title("Swiggy - Price Range Distribution")  # Set plot title
plt.xlabel("Price")  # Set x-axis label
plt.ylabel("Frequency")  # Set y-axis label
plt.show()  # Display the plot
```
